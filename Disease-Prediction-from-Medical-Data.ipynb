```python
#import libraries
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
# Load the dataset
Medical_data=pd.read_csv('Disease_Prediction_data.csv')
```

```python
#print the first 5 rows
Medical_data.head()
```

```python
# determine num of (rows,columns)
Medical_data.shape
```

```python
# check if there is null_values or not
Medical_data.isna().sum()
```

```python
#descriptive statistics for each column in the Medical_data DataFrame
Medical_data.describe()
```

```python
# convert data into numeric values
Encode_data=LabelEncoder()

# Perform label encoding on categorical columns
Medical_data['Sex']= Encode_data.fit_transform(Medical_data['Sex'])
Medical_data['ChestPainType']= Encode_data.fit_transform(Medical_data['ChestPainType'])
Medical_data['ExerciseAngina']= Encode_data.fit_transform(Medical_data['ExerciseAngina'])
Medical_data['ST_Slope']= Encode_data.fit_transform(Medical_data['ST_Slope'])
Medical_data['RestingECG']= Encode_data.fit_transform(Medical_data['RestingECG'])
```

```python
#print data after encoded to check that all categorized columns convert into numeric
Medical_data
```

```python
# Count the values in the 'HeartDisease' column
Medical_data['HeartDisease'].value_counts()
```

```python
# Calculate the mean values by 'HeartDisease' group
Medical_data.groupby('HeartDisease').mean()
```

```python
# Separate the features and target variables
X= Medical_data.drop(columns='HeartDisease' , axis=1)
Y=Medical_data['HeartDisease']
```

```python
# print X all features without target value
print(X)
```

```python
# print Y target value
print(Y)
```

```python
# Create an instance of StandardScaler
scaler= StandardScaler()
```

```python
# Fit the feature variables
scaler.fit(X)
```

```python
#transform the feature variables
standardized_data = scaler.transform(X)
```

```python
print(standardized_data)
```

```python
X=standardized_data
Y=Medical_data['HeartDisease']
```

```python
print(X)
print(Y)
```

```python
# Split the data into training and testing sets
X_train ,X_test ,Y_train ,Y_test= train_test_split( X,Y  ,test_size=0.2 ,stratify=Y ,random_state=42)
```

```python
print(X.shape, X_train.shape ,X_test.shape)
```

```python
classifier= svm.SVC(kernel='linear')
```

```python
#training the support vector Machine Classifier
classifier.fit(X_train, Y_train)
```

```python
#accuracy score on tranning_data

X_train_prediction=classifier.predict(X_train)
Tranning_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
#accuracy score on testing_data

X_test_prediction= classifier.predict(X_test)
Testing_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
#print accuracy of tranning_data & testing_data

print('Accuracy score of tranning data:',Tranning_data_accuracy)
print('Accuracy of Testing_data:',Testing_data_accuracy)
```
